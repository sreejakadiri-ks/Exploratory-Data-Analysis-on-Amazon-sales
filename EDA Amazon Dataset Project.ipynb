```python
#--------------------------------#install the necessary libraries needed for analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#------------------------------------# reading the dataset 
amadata = pd.read_csv("amazon.csv")
```

```python
#------------------------------# checking the memory, types on the dataset
amadata.info()
```

```python
#-----------------------------# to check if the dataset contains any null values
amadata.isna()
```

```python
#-------------------------------# checking the individual count of all the product category
amadata.nunique()
```

```python
#---------------------------------------# filling the nan values with sum() aggregation function
amadata.isna().sum()
```

```python
#--------------------------------------# to check nan values in specific column
amadata['discount_percentage'].isna()
```

```python
#-------------------------------------------# finding the average discount per product category
# first clean the discount_percentage column 
amadata['cleaned_discount_percentage'] = amadata['discount_percentage'].str.replace('%','').astype(float)
```

```python
# now finding the average discount based on cleaned discount percent per category
average_discount = amadata.groupby('category')['cleaned_discount_percentage'].mean().sort_values(ascending=False)
```

```python
print("The Average discount per product category is:\n",average_discount)
```

```python
#--------------------------------# to find the product which is high in discount but having low in rating 
# firstly clean the rating column
amadata["cleaned_rating"] = pd.to_numeric(amadata['rating'],errors="coerce")
# here we used errors='coerce' because to convert the non-convertible values(categorical) into the nan values
```

```python
# then from the cleaned discount, get percentage greater than 70%
high_discount = amadata[amadata['cleaned_discount_percentage']>70]
```

```python
# now finding the high in discount, but low in rating
high_discount_low_rating = high_discount.sort_values(by='cleaned_rating',ascending=True)
```

```python
# to view the top 5 products with high in discount but low in rating
print('The Products with High in Discount, but low in rating are:\n',high_discount_low_rating)
```

```python
#-----------------------------# to find the average diiference between actual price and discounted price 
# now cleaning the actual price attribite
amadata['cleaned_actual_price'] = amadata['actual_price'].str.replace('₹','').str.replace(',','').astype(float)
# now cleaning the discount price column
amadata['cleaned_discount_price'] = amadata['discounted_price'].str.replace('₹','').str.replace(",",'').astype(float)
```

```python
# to check the original price difference
amadata['price_difference'] = amadata['cleaned_actual_price']-amadata['cleaned_discount_price']
```

```python
# checking the average price difference
average_price_difference = amadata.groupby('category')['price_difference'].mean().sort_values(ascending=True)
#printing the result
print('The average price difference between actual price and discount price is:\n', average_price_difference)
```

```python
#--------------------------------------# to find the top 5  products and  their average rating
import pandas as pd
import matplotlib.pyplot as plt
cleaned_dataset = pd.read_csv('new_amadata.csv')
# Cleaning the rating column
cleaned_dataset['rating'] = pd.to_numeric(cleaned_dataset['rating'], errors='coerce')
# Group by product_name and get average rating
product_names = cleaned_dataset.groupby('product_name')['rating'].mean().sort_values(ascending=False).head(5)
# now ploting the graph 
plt.figure(figsize=(10, 6))
plt.bar(product_names.index, product_names.values, color='seagreen')
plt.xlabel('Product Name')
plt.ylabel('Average Rating')
plt.title('Top 5 Products and its Average Rating')
plt.xticks(rotation=90, ha='right')
plt.tight_layout() # we use tight_layout() to prevent the overlapping and checks that ploting can be done without disturbances.
plt.show()
```

```python
#--------------------------------------# Top 5 most reviewed products
# import libraries 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#load and clean the data
cleaned_dataset=pd.read_csv('new_amadata.csv')
cleaned_rating=pd.to_numeric(cleaned_dataset['rating'].str.replace(',',''),errors='coerce')
#droping the nan values in rating column
cleaned_rating=cleaned_dataset.dropna(subset=['cleaned_rating'])
#to get top 5 ratings
top_5_rating = cleaned_dataset.sort_values(by=['cleaned_rating'],ascending=False).head(5)
```

```python
# representing through scatterplot
sns.scatterplot(x='product_name',y='cleaned_rating',data=top_5_rating,color='black')
plt.xlabel('product names')
plt.ylabel('rating')
plt.title('To Represent the Top 5 most rated products')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
